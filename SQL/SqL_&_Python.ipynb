# Join two tabels and read them in a dataframe using read_sql_query

```python
query = '''
 SELECT Customers.CustomersID,Customers.CustormerName,Customers.ContactName,Products.Units,
 Products.Prices FROM Customers INNER JOIN Products ON Customers.CustomersID=Products.ProductID
'''
df = pd.read_sql_query(query,engine) # Inner Join
df
```

# Left Join

```python
query = '''
 SELECT Customers.CustomersID,Customers.CustormerName,Customers.ContactName,Products.Units,
 Products.Prices FROM Customers left JOIN Products ON Customers.CustomersID=Products.ProductID
'''
df = pd.read_sql_query(query,engine) # Inner Join
df
```

# Right Join

```python
query = '''
 SELECT Customers.CustomersID,Customers.CustormerName,Customers.ContactName,Products.Units,
 Products.Prices FROM Customers right JOIN Products ON Customers.CustomersID=Products.ProductID
'''
df = pd.read_sql_query(query,engine) # Inner Join
df
```
