# 4.1Create a table named ‘PRODUCT’ with following columns and constraints.



```python
cur.execute("create table Product(ProductCode int NOT NULL AUTO_INCREMENT,ProductName varchar(225), ProductPrice int(255), Manufacture varchar(255),PRIMARY KEY (ProductCode),UNIQUE(Manufacture))")
mydb.commit()
```

# 5.3 Select the name of the products with a price less than or equal to $200.

```python
query =''' select ProductName from Product where ProductPrice <= 200'''
qu3=pd.read_sql_query(query,engine)
qu3
```

# 5.4 Select all the products with a price between $60 and $120.

```python
query ='''select * from Product where ProductPrice between 60 and 120'''
qu4 = pd.read_sql_query(query,engine)
qu4 
```

# 5.5 Select the name and price in cents (i.e., the price must be multiplied by 100).

```python
query ='''select ProductName,ProductPrice*100  from Product'''
df5 = pd.read_sql_query(query,engine)
df5 
```

# 5.6 Compute the average price of all the products

```python
query = ''' select avg(ProductPrice) as Product_Price_of_Average from product'''
df6 = pd.read_sql_query(query,engine)
df6
```

# 5.7Compute the average price of all products with manufacturer code equal to 2.

```python
query = '''select ProductCode,avg(ProductPrice) as Product_Price_of_Average,Manufacture from Product where ProductCode = 2 '''
df7 = pd.read_sql_query(query,engine)
df7
```

# 5.8 Compute the number of products with a price larger than or equal to $180.

```python
query = ''' select count(*) as Product_Count from Product where ProductPrice >= 180'''
df8 = pd.read_sql_query(query,engine)
df8
```

# 5.9	Select the name and price of all products with a price larger than or equal to $180, and sort first by price (in descending order)!

```python
query = ''' select ProductName,ProductPrice from Product where ProductPrice>=180 order by ProductPrice desc'''
df9 = pd.read_sql_query(query,engine)
df9
```

# 5.10Select the average price of each manufacturer's products, showing the manufacturer's name.

```python
query ='''select avg(ProductPrice) as Average_ProductPrice,Manufacture from Product group by Manufacture'''
df10=pd.read_sql_query(query,engine)
df10
```

# 5.11Select the names of manufacturers whose products have an average price larger than or equal to $150.

```python
query =''' select Manufacture, avg(ProductPrice) as Average_ProductPrice from Product group by Manufacture having avg(ProductPrice)>=150 order by avg(ProductPrice) ASC  '''
df11=pd.read_sql_query(query,engine)
df11
```

# 5.12Select the name and price of the cheapest product.

```python
query='''select ProductName,ProductPrice as Cheapest_ProductPrice from Product where ProductPrice = (select min(ProductPrice) from Product)'''
df12 =pd.read_sql_query(query,engine)
df12
```
