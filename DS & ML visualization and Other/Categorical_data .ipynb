# Encoding Using sklearn

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'Marks':['78','30','50','92','80','66'],
                'Grade':['c','F','D','A','B','C']})
df
```

# A=0,B=1,C=2,D=3,F=4

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Grade'] = le.fit_transform(df['Grade'])
df
```

# One Hot Encoding

```python
df=pd.DataFrame({'Image':['Image1','Image2','Image3','Image4','Image5','Image6'],
                'Category':['cat','dog','cat','human','human','cat']})
df
```

```python
new_df=pd.get_dummies(df,columns=['Category'])
new_df
```

```python
df=pd.DataFrame({'Image':['Image1','Image2','Image3','Image4','Image5','Image6'],
                'Category':['cat','dog','cat','human','human',np.nan]})
df
```

```python
#pd.get_dummies() method ignore the NaN value while encoding.
new_df=pd.get_dummies(df,columns=['Category'])
new_df
```

# To consider the NaN value, use parameter dummy_na = True

```python
new_df=pd.get_dummies(df,columns=['Category'],dummy_na=True)
new_df
```
