```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('uk-election-results.csv')
```

```python
df
```

# Seaborn:line Plot

```python
sns.set(style="darkgrid")
sns.lineplot(data=df,x="year",y="conservative")
```

# Seaborn:Scatter Plot

```python
sns.lmplot(x='year',y='conservative', data=df)
```

```python
fig=sns.lmplot(x='year', y='conservative',data=df, fit_reg=False, hue='year')
```

```python
sns.set()
plt.figure(figsize=(8,6))
fig=sns.scatterplot(x='year',y=df["conservative"],data=df)
plt.xlabel('Conservative')
plt.ylabel('Seats')
```

# Seaborn: Bar Plot

# A) Single column

# Seaborn:barplot

```python
sns.set()
plt.figure(figsize=(12,8))
sns.barplot(x='year',y=df["others"],data=df)
plt.xticks(rotation=90)
plt.title('Bar Plot')
plt.xlabel('Others')
plt.ylabel('Seats')
```

```python
ax = sns.barplot(data=df, x="year", y="conservative")

ax.set_title('Uk Election Results')
ax.grid(color='#AAAAAA')
ax.set_ylabel('Seats')
ax.set_xlabel(None)
ax.set_xticklabels(df["year"].unique().astype(str),rotation=45);
```

# Seaborn:Histogram

```python
plt.figure(figsize=(8,6))
sns.distplot(df['others'],kde=False,color='red',bins=8)
```

# A) Multiple columns

```python
#Melt DataFrame
melted_df = pd.melt(df,id_vars=["year"],var_name="party")
melted_df
```

# Seaborn:Line Plot

```python
sns.set()
plt.figure(figsize=(12,8))
ax=sns.lineplot(data=melted_df,x="year",y="value",hue="party")
ax.set_title('UK Election Results',fontsize=18)
ax.grid(color='#FFFFFF')
ax.set_ylabel('Seats',fontsize=18)
ax.set_xlabel('Year',fontsize=18)
```

# Seaborn:Scatter Plot

```python
sns.set()
plt.figure(figsize=(12,8))
ax=sns.scatterplot(data=melted_df,x="year",y="value",hue="party")
ax.set_title('UK Election Results')
ax.grid(color='#DDDDDD')
ax.set_ylabel('Seats')
ax.set_xlabel('Year')
ax.legend(loc='upper left')
ax.set_xticklabels(labels=df['year'],rotation=45,ha='right');
plt.tick_params(axis='y',colors='red',rotation=45)
```

# Seaborn:Bar plot(A)

```python
sns.set()
plt.figure(figsize=(12,8))
sns.barplot(x='year',y='value',hue='party',data=melted_df)
plt.xticks(rotation=90)
plt.xlabel('year')
plt.ylabel('party')
plt.legend(loc='upper right')
```

# Seaborn:Bar Plot(b)

```python
sns.set()
plt.figure(figsize=(12,5))
ax=sns.barplot(data=melted_df,x="year",y="value",hue="party",
              palette=['blue','red','yellow','green'],
              saturation=0.8)
ax.set_title('UK Election Result')
ax.grid(color='#CCCCCC')
ax.set_ylabel('Seats')
ax.set_xlabel(None)
```

# Seaborn:Swarmplot

# Swarmplot gives a better picture in terms of the description of the relationship amongst categorial data variables.

```python
sns.set()
plt.figure(figsize=(12,8))
ax=sns.swarmplot(data=melted_df,x="year",y="value",hue="party",
                palette=['blue','red','yellow','green']
                )
ax.set_title('UK Election Results')
ax.grid(color='#CCCCCC')
ax.set_ylabel('seats')
ax.set_xlabel('Year')
```
