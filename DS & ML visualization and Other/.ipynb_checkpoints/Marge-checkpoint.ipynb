```python
import pandas as pd
```

# Creating Our own dataframe

```python
df1=pd.DataFrame({
    "state":["West Bengal","Rajasthan","Tamil Nadu","Maharashtra","Punjab"],
    "capital":["Kolkata","Jaipur","Chennai","Mumbai","Chandigarh"],
    "population":[334.67*10**6,75.3*10**6,82.5*10**6,66.9*10**6,255.5*10**6]
})
df1
```

```python
df2=pd.DataFrame({
    "state":["West Bengal","Rajasthan","Tamil Nadu","Maharashtra","Punjab"],
    "humidity":[82,68,75,45,67]
})
df2
```

# Merging both of them with a common field state

```python
df = pd.merge(df1,df2, on="state")
df
```

# Creating 2 Uneven dataframes

```python
df1=pd.DataFrame({
    "state":["West Bengal","Rajasthan","Tamil Nadu","Maharashtra","Punjab"],
    "capital":["Kolkata","Jaipur","Chennai","Mumbai","Chandigarh"],
    "population":[334.67*10**6,75.3*10**6,82.5*10**6,66.9*10**6,255.5*10**6]
})
df1
```

```python
df2 = pd.DataFrame({
    "state":["West Bengal","Haryana","Kashmir","Punjab"],
    "humidity":[65,68,71,45]
})
df2
```

# Inner Joning DataFrame

```python
df_inner = pd.merge(df1,df2,on="state",how="inner")
df_inner
```

# Outer Joning DataFrame

```python
df_outer = pd.merge(df1,df2,on="state",how="outer")
df_outer
```

# Left Joning DataFrame

```python
df_left = pd.merge(df1,df2,on="state",how="left")
df_left
```

# Right Joning DataFrame

```python
df_right = pd.merge(df1,df2,on="state",how="right")
df_right
```

# Outer Join With Indicator

```python
df3=pd.merge(df1,df2,on="state",how="outer",indicator=True)
df3
```

# Creating two dataframes with same columns

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38],
    "humidity":[65,68,71,75]
    
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","san diego"],
    "temperature":[21,14,35],
    "humidity":[65,68,71]
    
})
df2
```

# Distinguishing the left and right df columns

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

# Giving own custome suffix

```python
df=pd.merge(df1,df2,on="city",how="outer",suffixes=('_first','_second'))
df
```
