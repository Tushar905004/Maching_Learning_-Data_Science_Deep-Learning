# Q1.Handling missing data

1.Read the csv file with the column ‘day’ in date format.

```python
df=pd.read_csv("customer_data.csv",parse_dates=[0])
df
```

2.Make the day column the index

```python
df.set_index('day',inplace=True)
df
```

3.Drop all the rows with blank values in columns - ‘id’,’purchase_status’
and ‘name’

```python
df.dropna(subset=['id','name','purchase_status'],inplace=False)
```

4.Drop the rows with all the rows blank

```python
df.dropna(axis=0,how='any',inplace=False)
```

5.In the columns - age,salary (replace the blank values with NaN)

```python
df1=df.replace({'age':0,'salary':0},np.NaN)
df1
```

6.In salary column - remove phrases like ‘dollars’ or ‘INR’ using regex

```python
df1=df.replace({'salary':'[dollars,in]'},'',regex=True)
df1
```

7.In the column - age and salary, interpolate the NaN values

```python
df1['salary']=pd.to_numeric(df1['salary']).fillna(0).astype(float)
df1['salary'].dtypes
#df2=df1.interpolate(df1=['age','salary'],method='time')
#df2
```

```python
df2=df1.replace({'salary':0.0},np.NaN)
df2
```

```python
df2.dtypes
```

```python
new=df2.interpolate(df2=['age','salary'],method="time",axis=0)
new
```

8.Create the index for missing dates in the dataframe, and fill the missing
values in a suitable form.

```python
date =pd.date_range(start="2020-01-01",end="2020-12-01")
index=pd.DatetimeIndex(date)
index
```

```python
df3=df.reindex(index)
df3
```

9.In the column - age and salary, operate ‘ffill’ and ‘bfill’ the NaN
values(Results has to be saved on a separate dataframe)

```python
df4=df3.fillna(method='ffill')
df4
```

```python
df5=df3.fillna(method='bfill')
df5
```

```python
df6=df5.interpolate(df3=['age','salary'],method='time')
df6.head(20)
```

3.On a separate data frame perform outer,left,right join on id.

```python
#outer:
mer=pd.merge(df,join1,on="id",how='outer',indicator=True)
mer
```

```python
#left:
mer=pd.merge(df,join1,on="id",how='left',indicator=True)
mer
```
