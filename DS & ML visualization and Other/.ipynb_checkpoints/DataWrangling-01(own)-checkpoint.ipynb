# Data Preprocessing

3.Dealing with Missing Data

3.1 Exploring(Determining) the missing data

3.1.1 Heatmap

```python
cols = df.columns[:10]#first 10 columns
colours = ['#ffff00','#000000']# specify the colours -yellow is missing..black is not missing.
sns.heatmap(df[cols].isnull(),cmap=sns.color_palette(colours))
```

# 3.1.2 Percentage List

```python
#if it's a larger data and the visualization takes too long can do this.
# % of missing

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

# 3.1.3 Histogram

```python
#first create missing indicator for features with missing data

for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing>0:
        print('Created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)]=missing
        #print(missing)
#then based on the indicator, plot the histogram of missing values:

ismissing_cols=[col for col in df.columns if 'ismissing' in col]
df['num_missing']=df[ismissing_cols].sum(axis=1)
df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index',y='num_missing')
```

# 3.2 Recommended Actions

# 3.2.1 Drop the Observation

```python
#drop rows with a lot of missing values:
ind_missing = df[df['num_missing']>30].index
df_less_missing_rows=df.drop(ind_missing,axis=0)
```

```python
print(df_less_missing_rows.shape)
```

```python
#if it's a larger dataset and the visualization takes too long can do this:
#% of missing
for col in df_less_missing_rows.columns:
    pct_missing=np.mean(df_less_missing_rows[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```
