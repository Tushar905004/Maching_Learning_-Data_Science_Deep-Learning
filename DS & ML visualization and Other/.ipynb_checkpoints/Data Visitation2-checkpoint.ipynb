# 5.3.2 More customized x/y Tricks

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.random.random(10)
plt.plot(x,y)

#Custom ticks
plt.tick_params(axis='x',color='red',direction='out',length=10,width=5)
plt.tick_params(axis='y',color='green',rotation=45)
```

# 5.4 Customizing the line plot with marker and line styles

```python
x=np.arange(1,11)
y=np.random.random(10)
fig,ax=plt.subplots(figsize=(10,6))
plt.tick_params(axis='x',color='red',direction='out',length=10,width=5)
plt.tick_params(axis='y',color='green',rotation=150,length=10,width=4)
ax.plot(x,y,marker='o',markerfacecolor='blue',markersize=12,linestyle='--',color='skyblue',linewidth=4)
```

# 5.5 customizing line plots with Annotations

Annotating each point with its(x,y)coordinates.we need to specify the coordinates at which the annotations should appear.

```python
fig,ax=plt.subplots(figsize=(10,4))
x=np.arange(1,11)
y=np.random.random(10)
#y=[2,0.8,1.4,3,0.67,2.2,2.5,4,6,3]
ax.plot(x,y,marker='o',linestyle='--')
for i in range(len(x)):
    xi="{:.1f}".format(x[i])
    yi="{:.1f}".format(y[i])
    s=str("p("+str(xi)+ ',' + str(yi)+')')
    ax.text(x[i]+ 0.03, y[i]+0.03,s)
```

# 6. Multiple line plots

6.1 Multiple line plots with legends

```python
fig, ax=plt.subplots(1,1, figsize=(8,4))
x=np.linspace(1,10,50)
y=np.sin(x)
y1=np.cos(x)
ax.plot(x,y,label='Sine Func')
ax.plot(x,y1,label='cos Func')
ax.set_title('Multiple line plots with legends')
ax.legend()
```

6.1.1 Customiziing the line plots

```python
fig, ax=plt.subplots(1,1,figsize=(8,4))
x=np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
ax.plot(x, y1, linestyle=':', color='olive', label='sine Func')
ax.plot(x, y2, color='#FF0000', label='Cos Func', linewidth=2)
ax.set_title('Multiple line plots with legends')
ax.legend()
```

6.1.2 Customizing the legend

```python
fig, ax = plt.subplots(1,1,figsize=(8,4))
x=np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
ax.plot(x, y1, linestyle=':',color='olive',label='Cos Func', alpha=0.5)
ax.plot(x, y2, color='#00FF00',label='Sine Func', linewidth=2)
ax.set_title('Multiple line plots with legends')
ax.legend(fancybox=True,loc='upper left',framealpha=1,shadow=True,borderpad=0.5,fontsize=10)
```

6.1.3 How to use seabron darkgrid/Customizing look and feel with style

```python
plt.style.use('seaborn-darkgrid')
fig,ax=plt.subplots(1,1,figsize=(8,4))
x=np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
ax.plot(x,y1,linestyle=':',color='olive',label='Cos Func',alpha=0.3)
ax.plot(x,y2,color='#00FF00',linewidth=2,label='Sine Func')
ax.set_title('Multiple line plots with legends')
ax.legend()
```

# 8.Adding a second x/y-axis in a graph

```python
w=4
h=3
d=70
t=np.arange(0.01,10.0,0.01)
exp_data=np.exp(t)
sin_data=np.sin(2*np.pi*t)

plt.figure(figsize=(w,h),dpi=d)
ax=plt.gca()#get the plot axes object

ax.plot(t,exp_data)
ax.set_title('Exponeotial Plot')
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
```

```python
fig, ax1=plt.subplots()
color='tab:red'
ax1.set_xlabel('Time(s)')
ax1.set_ylabel('Exponential',color=color)
ax1.plot(t,exp_data,color=color)
ax1.tick_params(axis='y',labelcolor=color)

ax2 = ax1.twinx()#Instantiate a second axes that shares the same x-axis(similarly twiny() to share the same y axis)

color = 'tab:blue'
ax2.set_ylabel('Sine Func',color=color)# we already handled the x-label with ax1
ax2.plot(t,sin_data,color=color)
ax2.tick_params(axis='y',labelcolor=color)

ax2.yaxis.grid()
```

```python
fig, ax=plt.subplots(1,1, facecolor='lightgray')
ax.axis([0,10,0,10])

#transfore=ax.transData is the default, but we'll specify it anyway
ax.text(1,5, ". Data:(1,5)", transform=ax.transData)
ax.text(0.5,0.1, " . Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(1,1, ". Figure:(1,1)", transform=fig.transFigure)
```

```python
fig, ax = plt.subplots()
x=np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local maximum',xy=(6.28,1),xytext=(10,5),
            arrowprops=dict(facecolor='black',shrink=0.05))                                    
ax.annotate('local minimum',xy=(5*np.pi,-1),xytext=(2,-6),
           arrowprops=dict(arrowstyle="->", connectionstyle="angle3,angleA=-50,angleB=0"))
```
