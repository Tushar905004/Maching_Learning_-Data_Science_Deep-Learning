```python
import pandas as pd
```

# Reading the CSV Files

```python
df1=pd.read_csv('india_weather.csv')
df2=pd.read_csv('usa_weather.csv')
```

```python
df1
```

```python
df2
```

# Joining them using concat

```python
df=pd.concat([df1,df2])#more than 2 dataframes can be concatenated together
df
```

# ignoring the index

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

# Adding Keys

```python
df=pd.concat([df1,df2],keys=["India","US"])#Won't work when ignore_index=True
df
```

```python
df.loc["India"]#accessing the keys
```

```python
df.loc["US"]#accessing the keys
```

# Column wise concatination

```python
#making another dataframe
windspeed_df = pd.DataFrame({
    "City":["Delhi","Kolkata","Chennai","Mumbai"],
    "Windspeed":[7,12,10,7]
})

windspeed_df
```

```python
df1
```

```python
df=pd.concat([df1,windspeed_df],axis=1)#axis=1 means column wise concat, default value is 0
df
```

# When data is not arranged sequentially

```python
df1
```

```python
windspeed_df2 = pd.DataFrame({
    "city":["Delhi","Pune","Jaipur","Mumbai"],
    "windspeed":[7,12,10,7]
})
windspeed_df2
```

```python
df=pd.concat([df1,windspeed_df2],axis=1)
df
```

# Menualy Indexing Set

```python
windspeed_df2 = pd.DataFrame({
    "city": ["Delhi","Pune","Jaipur","Mumbai"],
    "windspeed": [7,12,10,7]
}, index =[0,6,7,3]
)

windspeed_df2
```

```python
df=pd.concat([df1,windspeed_df2],axis=1)
df
```

# Concatinating a Panda Series

```python
st=pd.Series(["Delhi","west bengal","Tamil Nadu","Maharashtra","Karnataka","Andra Pradesh","Maharashtra","Rajasthan","Punjab"],name="State")
st
```

```python
df=pd.concat([df1,st],axis=1)
df
```
