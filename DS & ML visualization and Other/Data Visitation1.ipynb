```python
fig7, f7_axs = plt.subplots(ncols=2, nrows=2)
gs = f7_axs[0, 1].get_gridspec()
# remove the underlying axes
for ax in f7_axs[0:, -1]:
    ax.remove()
axbig = fig7.add_subplot(gs[0:, -1])
axbig.annotate('Big Axes \nGridSpec[0:, -1]', (0.1, 0.5),
               xycoords='axes fraction', va='center')

#fig7.tight_layout()
```

```python
#import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)
print(x)
y=np.random.random(10)
print(y)
fig,ax=plt.subplots(figsize=(10,6))
ax.grid()
ax.plot(x,y)
```

2.1.2 ploting a sinusiodal function

```python
x=np.linspace(1,10,50)
#print(x)
y=np.sin(x)
fig,ax=plt.subplots(figsize=(10,6))
ax.grid()
ax.plot(x,y)
```

2.1.3 ploting a sinusiodal function without creating any ax object

```python
x=np.linspace(1,10,50)
y=np.sin(x)
fig,ax=plt.subplots(figsize=(7,5))
plt.plot(x,y)
plt.grid()
ax.set(title='Sine Function',
       xlabel='Time',
       ylabel='Amplitude')
```

# 3.Matplotlib: Creating scatter plots

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.random.random(10)
print(x)
print(y)
fig,ax=plt.subplots(figsize=(10,6))

ax.scatter(x,y)
```

# 4. Matplotlib: Creating bar plots

4.1.1 Example:ploting 10 data points generated randomly

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.random.random(10)
print(x)
print(y)
fig,ax=plt.subplots(figsize=(10,6))
ax.bar(x,y)
```
