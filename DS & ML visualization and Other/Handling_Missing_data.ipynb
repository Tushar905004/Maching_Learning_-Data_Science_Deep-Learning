# Converting day as date[parse_dates]

```python
df=pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
type(df.day[0])
```

# Setting data as index

```python
df.set_index('day',inplace=True)#inplace = True allows changes in the present dataframe
df
```

# Filling all NaN values with 0 using Fillna()

```python
new_df=df.fillna(0)
new_df#0 in events don't make sence
```

```python
new_df = df.fillna({'temp':0,
                    'windspeed':0,
                     'event':'No Event'})
new_df #now 0 temperature indicates huge temperature change
```

# Instead of replacing with 0, do a forward fill

```python
new_df=df.fillna(method="ffill")
new_df
```

```python
new_df=df.fillna(method='bfill')#backword fill
new_df
```

```python
new_df = df.fillna(method="bfill",axis="columns")#axis is either "index" or columns
new_df
```

# Instead of copying to all the rows, let the limit be 1

```python
new_df = df.fillna(method="ffill",limit=2)
new_df
```

# Instead of copying, try to calculate the missing values

```python
new_df=df.interpolate() #default linear, but can take many other methods as well like,ploynomial,spline,cubic,quadratic,etc...
new_df
```

# Now making the guess more on time bound

```python
new_df=df.interpolate(method="time")
new_df
```

# Now instead of filling values, drop the rows with the missing values.

```python
new_df=df.dropna()
new_df
```

# Instead of removing all na values, remove only the rows which has all na values, except of index.

```python
new_df=df.dropna(how='all')
new_df
```

# Setting threshold to 1,means deleting all rows with na values,except for those who have atleast 1 value

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
new_df = df.dropna(thresh=2)#atleast 2 values needed
new_df
```

```python
df
```

# Adding on the missing index value

```python
dt = pd.date_range("01-01-2017","01-11-2017")#date time format conversion
#print(dt)
idx=pd.DatetimeIndex(dt)#returning the list of dates within the date range
idx
```

```python
df = df.reindex(idx)
df
```

```python
new_df = df.interpolate(method="time")#interpolating the new indexed dataframe on temp, windspend
new_df
```

```python
new_df = new_df.fillna(method="ffill")#forward filling the events on the interpolated df
new_df
```

# Filling misc values with NaN, using replace function

```python
import pandas as pd
import numpy as np
df=pd.read_csv("weather.csv")
df
```

# Replacing single values

```python
new_df=df.replace(-99999, value=np.NaN)
new_df
```

# Pass a list of replacable values

```python
new_df=df.replace(to_replace=[-99999,-88888,88888],value=np.NaN)
new_df
```

# Replacing the zeroes as well, for pressure and event

```python
new_df= df.replace(to_replace=[-99999,-88888,88888,0,'0'],value=np.NaN)
new_df
```

```python
df
```

# Replacing column wise

```python
new_df = df.replace({
        'temp':-99999,
        'dewpoint':-99999,
        'humidity':[88888,-88888],
        'sea_level_pressure':0.00,
        'event':'0'
},np.NaN)
new_df
```

# Replacing valuewise

```python
new_df=df.replace({-99999:np.NaN,-88888:0})
new_df
```

# Replacing Regular Expression

```python
df1 = pd.DataFrame({
    'temp':['32f',45,58,32,55,23,'30F'],
    'windspeed':[1,2,4,'6mph','7mph',9,7]
})
df1
```

```python
#when windspeed is 6mph,7mph,etc. & temperature is 32F,28F etc
new_df = df1.replace({'temp':'[A-Za-z]','windspeed':'[a-z]'},'',regex=True)
new_df
```
