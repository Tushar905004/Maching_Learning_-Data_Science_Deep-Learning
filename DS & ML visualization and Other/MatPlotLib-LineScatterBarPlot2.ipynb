### 5.5 Customizing line plots with Annotations

#### Annotating each point with its (x,y) coordinates. We need to specify the coordinates at which the annotations should appear.

```python
fig,ax=plt.subplots(figsize = (10,4))
x = np.arange(1,11)
y = [2,0.8,1.4,3,0.67,2.2,2.5,3,-2,5]
print(x)
print(y)
ax.plot(x,y, marker = '*', linestyle = '--')
for i in range(len(x)):
    xi = "{:.1f}".format(x[i])
    yi = "{:.1f}".format(y[i])
    s = str("P(" + str(xi) + ',' + str(yi) + ')' )
    ax.text(x[i] + 0.05, y[i] + 0.05, s)
```

# 6. Multiple line plots

### 6.1 Multiple line plots with Legends

```python
fig, ax = plt.subplots(1,1, figsize = (8,4))
x = np.linspace(1,10,10)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = [2,0.8,1.4,3,0.67,2.2,2.5,3,-2,5]
ax.plot(x, y1, label = 'Sine Func')
ax.plot(x, y2, label = 'Cos Func')
ax.plot(x, y3, label = 'Other Func')
ax.set_title('Multiple line plots with legends')
ax.legend()
#ax.legend(loc='upper left', frameon=False)
```

### 6.1.1 Customizing the line plots

```python
fig, ax = plt.subplots(1,1, figsize = (8,4))
x = np.linspace(1,10,50)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, linestyle = ':', color='olive', label = 'Sine Func')
ax.plot(x, y2, color='#2A507F', linewidth=2, label = 'Cos Func')
ax.set_title('Multiple line plots with legends')
ax.legend()
```

### 6.1.2 Customizing the Legend

```python
fig, ax = plt.subplots(1,1, figsize = (8,4))
x = np.linspace(1,10,50)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, linestyle = ':', color='olive', label = 'Cos Func',alpha=0.75) # alpha indicates transparency (0=transparent, 1=opaque)
ax.plot(x, y2, color='#00FF00', linewidth=2, label = 'Sine Func')
ax.set_title('Multiple line plots with legends')
#ax.legend(loc='upper left', frameon=False)
#ax.legend(frameon=False, loc='lower center', ncol=2)
ax.legend(fancybox=True, loc='upper left', framealpha=0.5, shadow=True, borderpad=1, fontsize=10,)
```

# 9. Adding a second x/y-axis in a graph

```python
# USE matplotlib.axes._base._AxesBase.twiny() TO ADD A SECOND X-AXIS IN A MATPLOTLIB GRAPH
```

```python
# Use matplotlib.pyplot.gca() to get the current axes of the graph. 
# Call matplotlib.axes._base._AxesBase.twiny() on the current axes to create a twin of axes that have a 
# shared y-axis with the current axes. Call matplotlib.axes._axes.Axes.set_xticks(ticks) with ticks as a list 
# of floats to set the x-ticks on the twin axes. Call matplotlib.axes._axes.Axes.set_xlabel(xlabel) with xlabel 
# as the label for the x-axis for the original axes and twin axes.
```

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#import random

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
exp_data = np.exp(t)
sin_data = np.sin(2 * np.pi * t)
```

```python
import matplotlib.pyplot as plt
w = 4
h = 3
d = 70
plt.figure(figsize=(w, h), dpi=d)
ax = plt.gca()  #Get the plot axes object

ax.plot(t,exp_data)

ax.set_title('Exponential Plot')
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
```

```python
%matplotlib notebook

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Exponential', color=color)
ax1.plot(t, exp_data, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis (Similarly twiny()to share the same y axis)

color = 'tab:blue'
ax2.set_ylabel('Sine Func.', color=color)  # we already handled the x-label with ax1
ax2.plot(t, sin_data, color=color)
ax2.tick_params(axis='y', labelcolor=color)


ax2.yaxis.grid()
```

# 10. Anonating a plot 

```python
fig, ax = plt.subplots(1,1, facecolor='lightgray')
ax.axis([0, 10, 0, 10])

# transform=ax.transData is the default, but we'll specify it anyway
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(1, 1, ". Figure: (1, 1)", transform=fig.transFigure);
```

```python
# The arrow style is controlled through the arrowprops dictionary, which has numerous options available.
```

```python
%matplotlib notebook

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=-90,angleB=0"));
```
