```python
label = df.target_class.values
df.drop(["target_class"],axis=1,inplace=True)
features=df.values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,2))
features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_scaled,label,test_size=0.2)
```

# ** Support Vector Machine**

```python
#kernel="linear"
from sklearn import metrics
from sklearn.svm import SVC
svm_model = SVC( C=100,gamma=0.1, kernel="linear")
svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
#kernel="poly"
svm_model = SVC( C=200,gamma=0.2, kernel="poly")
svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
#kernel="sigmoid"
svm_model = SVC( C=150,gamma=0.3, kernel="sigmoid")
svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
