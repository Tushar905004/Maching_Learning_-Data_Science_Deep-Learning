```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Home_prices_USA.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)',fontsize=20, color='red')
plt.ylabel('price(us$)',fontsize=20,color='red')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.predict([[3300]]) #3300 Squre fit
```

```python
reg.intercept_
```

```python
#y=m*area+b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)',fontsize=20, color='red')
plt.ylabel('price(us$)',fontsize=20,color='red')
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='green')
```
